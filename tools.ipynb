# Tools :
Models can request to call tools that perform tasks such as fetching data from a database, searching the web, or running code. Tools are pairings of:
1. A schema, including the name of the tool, a desciption, and/or argument definitions (often a JSON schema).
2. A function or coroutine to execute.

```python
import os
from langchain.chat_models import init_chat_model

ollama_model = init_chat_model(
"ollama:llama3.1"
)
```

```python
response  = ollama_model.invoke("how is Michael Jordon is best in NBA history?")
```

```python
response.content
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str) -> str:
    """ Get the weather at a location"""
    return f"Its the sunny in {location}"


model_Tool = ollama_model.bind_tools([get_weather])


```

```python
response = model_Tool.invoke("whats the weather like in Boston?")
print(response)

for tool_Call in response.tool_calls:
    ## VIew tool calls made by the model
    print(f"Tools : {tool_Call['name']}")
    print(f"Args: {tool_Call['args']}")
```
